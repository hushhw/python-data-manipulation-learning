```python
import matplotlib.pyplot as plt
import numpy as np
```

使用 `np.linspace` 定义 x：范围是 (-3, 3)，个数是50。
     
曲线 1 用 (x, y1) 表示，曲线 2 用 (x, y2) 表示。

```python
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2
```

matplotlib 的 `figure` 就是一个单独的 figure 小窗口，小窗口里面可以有更多的小图片。

使用 `plt.figure` 定义一个图像窗口，使用 `plt.plot` 画曲线 1。

```python
plt.figure()
plt.plot(x, y1)
plt.show()
```

使用 `plt.figure` 定义一个图像窗口：编号为 3，大小为 (8, 5)，使用 `plt.plot` 绘制曲线 2。
      
把曲线 1 设置新样式：颜色设置为红色，曲线宽度为 1.0，类型为虚线。

```python
plt.figure(num=3, figsize=(8, 5))
plt.plot(x,y2)
plt.plot(x, y1, color='red', linewidth=5.0, linestyle='--')
plt.show()
```
