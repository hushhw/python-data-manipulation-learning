```python
import matplotlib.pyplot as plt
import numpy as np
```

生成 1024 个呈标准正态分布的二维数据组（N(0,1)）作为一个数据集，并图像化这个数据集。每一个点的颜色用 `T` 表示：

```python
n = 1024
X = np.random.normal(0, 1, n)
Y = np.random.normal(0, 1, n)
T = np.arctan2(Y, X)

plt.scatter(X, Y, s=75, c=T, alpha=.3)

plt.xlim(-1.5, 1.5)
plt.xticks(())
plt.ylim(-1.5, 1.5)
plt.yticks(())

plt.show()
```

使用 `plt.scatter` 来生成散点图   
输入 X 和 Y 作为 location，size=75，颜色为 T，color map 用默认值，透明度 alpha  为 50%。 x 轴显示范围定位(-1.5，1.5)，并用 xtick() 函数来隐藏 x 坐标轴，y 轴
